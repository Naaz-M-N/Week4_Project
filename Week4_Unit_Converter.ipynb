```python
import time

def print_letter_by_letter(message):
    for char in message:
        print(char, end='', flush=True)
        time.sleep(0.05)  # Adjust the sleep time for the desired speed
    print()

def meters_to_feet(meters):
    return meters * 3.28084

def feet_to_meters(feet):
    return feet / 3.28084

def celsius_to_fahrenheit(celsius):
    return celsius * 9/5 + 32

def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

def kg_to_pounds(kg):
    return kg * 2.20462

def pounds_to_kg(pounds):
    return pounds / 2.20462

def length_converter(value):
    for _ in range(2):  # Give the user two chances
        print("Select the unit to convert:")
        print("1. Meters to Feet")
        print("2. Feet to Meters")

        try:
            choice = int(input("Enter your choice (1/2): "))
            
            if choice == 1:
                result = meters_to_feet(value)
                print(f"{value} meters is equal to {result:.2f} feet.")
                return
            elif choice == 2:
                result = feet_to_meters(value)
                print(f"{value} feet is equal to {result:.2f} meters.")
                return
            else:
                print("Invalid choice. Please enter 1 or 2.")
        except ValueError:
            print("Invalid input. Please enter a numeric value.")
    
    print("You've exceeded the maximum number of attempts. Exiting.")

def temperature_converter(value):
    for _ in range(2):  # Give the user two chances
        print("Select the unit to convert:")
        print("1. Celsius to Fahrenheit")
        print("2. Fahrenheit to Celsius")

        try:
            choice = int(input("Enter your choice (1/2): "))
            
            if choice == 1:
                result = celsius_to_fahrenheit(value)
                print(f"{value} Celsius is equal to {result:.2f} Fahrenheit.")
                return
            elif choice == 2:
                result = fahrenheit_to_celsius(value)
                print(f"{value} Fahrenheit is equal to {result:.2f} Celsius.")
                return
            else:
                print("Invalid choice. Please enter 1 or 2.")
        except ValueError:
            print("Invalid input. Please enter a numeric value.")
    
    print("You've exceeded the maximum number of attempts. Exiting.")

def weight_converter(value):
    for _ in range(2):  # Give the user two chances
        print("Select the unit to convert:")
        print("1. Kilograms to Pounds")
        print("2. Pounds to Kilograms")

        try:
            choice = int(input("Enter your choice (1/2): "))
            
            if choice == 1:
                result = kg_to_pounds(value)
                print(f"{value} kilograms is equal to {result:.2f} pounds.")
                return
            elif choice == 2:
                result = pounds_to_kg(value)
                print(f"{value} pounds is equal to {result:.2f} kilograms.")
                return
            else:
                print("Invalid choice. Please enter 1 or 2.")
        except ValueError:
            print("Invalid input. Please enter a numeric value.")
    
    print("You've exceeded the maximum number of attempts. Exiting.")

def main():
    welcome_message = "Welcome to the Unit Converter!"
    print_letter_by_letter(welcome_message)

    try:
        value = float(input("Enter the value to convert: "))
        print("Select the type of converter:")
        print("1. Length Converter")
        print("2. Temperature Converter")
        print("3. Weight Converter")

        choice = int(input("Enter your choice (1/2/3): "))

        if choice == 1:
            length_converter(value)
        elif choice == 2:
            temperature_converter(value)
        elif choice == 3:
            weight_converter(value)
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

    except ValueError:
        print("Invalid input. Please enter a numeric value.")

if __name__ == "__main__":
    main()
```
